```python
import tensorflow as tf

# Reset the default TensorFlow graph
tf.compat.v1.reset_default_graph()

# Define input function
def input_fn(features, labels, batch_size, num_epochs, shuffle):
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(features))

    dataset = dataset.batch(batch_size).repeat(num_epochs)

    return dataset


# Define feature columns
feature_columns = [
    tf.feature_column.numeric_column(key=feature_name) for feature_name in X_Train.columns
]

# Create a LinearClassifier
linear_classifier = tf.estimator.LinearClassifier(
    feature_columns=feature_columns,
    n_classes=2,  # Number of classes for binary classification
    model_dir='your_model_directory'  # Choose or create a directory for saving model checkpoints
)

# Use the input function
batch_size = 40
num_epochs = 1000
shuffle = True

train_input_func = lambda: input_fn(X_Train, Y_Train, batch_size, num_epochs, shuffle)

# Train the LinearClassifier using the input function
model = linear_classifier.train(input_fn=train_input_func)

```
